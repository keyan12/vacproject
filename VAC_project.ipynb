```python

import os
import warnings
warnings.simplefilter('ignore')
```

```python

import numpy as np
import pandas as pd
```

```python
from google.colab.patches import cv_imshow
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

from skimage.io import imread, imshow
from skimage.transform import resize
from skimage.color import rgb2gray
```

```python
bikes=os.listdir("/content/drive/MyDrive/VAC dataset/bikes")
```

```python
cars=os.listdir("/content/drive/MyDrive/VAC dataset/Cars")
```

```python
limit=10
bikes_images=[None]*limit
j=0
for i in bikes:
    if(j<limit):
        bikes_images[j]=imread("/content/drive/MyDrive/VAC dataset/bikes/"+i)
        j+=1
    else:
        break
```

```python
cars_images=[None]*limit
j=0
for i in cars:
    if(j<limit):
        cars_images[j]=imread("/content/drive/MyDrive/VAC dataset/Cars/"+i)
        j+=1
    else:
        break
```

```python
imshow(bikes_images[1])
```

```python
imshow(cars_images[1])
```

```python
bikes_gray=[None]*limit
j=0
for i in bikes:
    if(j<limit):
         bikes_gray[j]=rgb2gray(bikes_images[j])
         j+=1
    else:
         break
```

```python
cars_gray=[None]*limit
j=0
for i in cars:
    if(j<limit):
        cars_gray[j]=rgb2gray(cars_images[j])
        j+=1
    else:
         break
```

```python
imshow(bikes_gray[1])
```

```python
imshow(cars_gray[1])
```

```python
bikes_gray[2].shape
```

```python
for j in range (10):
 rk=bikes_gray[j]
 bikes_gray[j]=resize(rk,(512,512))
```

```python
cars_gray[1].shape
```

```python
for j in range (10):
 rk=cars_gray[j]
 cars_gray[j]=resize(rk,(512,512))
```

```python
len_of_images_bikes=len(bikes_gray)
len_of_images_bikes
```

```python
len_of_images_cars=len(cars_gray)
```

```python
image_size_bikes=bikes_gray[1].shape
```

```python
image_size_cars=cars_gray[1].shape
```

```python
image_size_bikes
```

```python
image_size_cars
```

```python
flatten_size_bikes=image_size_bikes[0]*image_size_bikes[0]
```

```python
flatten_size_cars=image_size_cars[0]*image_size_cars[1]
```

```python
flatten_size_bikes
```

```python
flatten_size_cars
```

```python
for i in range(len_of_images_bikes):
  bikes_gray[i]=np.ndarray.flatten(bikes_gray[i]).reshape(flatten_size_bikes,1)
```

```python
for i in range(len_of_images_cars):
 cars_gray[i]=np.ndarray.flatten(cars_gray[i]).reshape(flatten_size_cars,1)
```

```python
bikes_gray=np.dstack(bikes_gray)
bikes_gray.shape
```

```python
cars_gray=np.dstack(cars_gray)
cars_gray.shape
```

```python
bikes_gray=np.rollaxis(bikes_gray,axis=2,start=0)
```

```python
cars_gray=np.rollaxis(cars_gray,axis=2,start=0)
```

```python
bikes_gray.shape
```

```python
cars_gray.shape
```

```python
bikes_gray=bikes_gray.reshape(len_of_images_bikes,flatten_size_bikes)
```

```python
cars_gray=cars_gray.reshape(len_of_images_cars,flatten_size_cars)
```

```python
bikes_gray.shape
```

```python
cars_gray.shape
```

```python
bikes_data=pd.DataFrame(bikes_gray)
```

```python
cars_data=pd.DataFrame(cars_gray)
```

```python

bikes_data
```

```python

cars_data
```

```python
bikes_data["label"]="bikes"
```

```python
cars_data["label"]="cars"
```

```python
bikes_data
```

```python
cars_data
```

```python
vehicle=pd.concat([bikes_data,cars_data])
```

```python
vehicle
```

```python
from sklearn.utils import shuffle
```

```python
transport_indexed=shuffle(vehicle).reset_index()
```

```python
transport_indexed
```

```python
vehicles=transport_indexed.drop(['index'],axis=1)
```

```python
vehicles
```

```python
vehicles.to_csv("/content/drive/MyDrive/VAC dataset/vehicles.csv")
```

```python
x=vehicles.values[:,:-1]
```

```python

y=vehicles.values[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC()
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python

y_pred
```

```python

for i in (np.random.randint(1,3,4)):
  predicted_images = (np.reshape(x_test[i], (512,512)).astype(np.float64))
  plt.title('predicted label: {0}'. format(y_pred[i]))
  plt.imshow(predicted_images, interpolation='nearest', cmap='gray')
  plt.show()
```

```python
from sklearn import metrics
```

```python
accuracy=metrics.accuracy_score(y_test,y_pred)
```
