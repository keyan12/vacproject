```python
import matplotlib.pyplot as plt
x = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
y = [15, 20, 25, 30, 35, 40]
plt.bar(x, y)
plt.title('Sales by Month')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.show()
```
